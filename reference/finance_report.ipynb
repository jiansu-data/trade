```python
def get_finance_data(chrome,stock_id):
    sid_input = chrome.find_element_by_xpath('//*[@id="sid_mapping"]')
    sid_input.send_keys(stock_id)
    time.sleep(0.2)
    chrome.find_element_by_xpath('//*[@id="sidoverview"]/button').click()
    time.sleep(10)
    for e in range(5):
        chrome.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)
    rep = chrome.find_elements_by_xpath('//*[@id="basec"]')
    all_data = rep[0].text
    for e in ['分類 ','獲利\n能力\n','償債\n能力\n','經營\n能力\n','現金\n流量\n','財務\n結構\n']:
        all_data =all_data.replace(e,'')
    for e in ['毛利率','營益率','淨利率','ROE','ROA','本業','負債比','長期資金佔\n固定資產比例','利息保障\n倍數','流動比','速動比','應收帳款\n周轉天數(天)','存貨週轉\n天數(天)','自由現金流(佰萬)','營運現金流(佰萬)','每股盈餘']:
        all_data = all_data.replace(e,"<<<<<<"+e)
    all_data
    data_dic = {}
    for e in data_lines2:
        _ = (e.split(' '))
        data_dic[_[0]] = _[1:]
    del data_dic['本業獲利率(%)']
    data_dic
    import pandas as pd
    df = pd.DataFrame(data_dic)
    return df
```
